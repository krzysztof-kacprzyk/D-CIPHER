```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import pandas as pd

file = glob.glob('../results/Figure_8_9/*.csv')[0]
df = pd.read_csv(file)

e_means = df['error_mean']
e_i_low = df['error_mean'] - df['error_std']
e_i_high = df['error_mean'] + df['error_std']

linewidth = 2.0
markersize = 10.0
xlabel_fontsize = 20
ylabel_fontsize = 20
title_fontsize = 20
legend_fontsize = 20
ticks_fontsize = 20
CB_color_cycle = ['#377eb8', '#ff7f00', '#4daf4a',
                  '#f781bf', '#a65628', '#984ea3',
                  '#999999', '#e41a1c', '#dede00']
fig, ax = plt.subplots(1, 1,figsize=(12,5),dpi=150)
ax.plot(np.arange(2,12),np.array(e_means),'o-',label='D-CIPHER',color=CB_color_cycle[0],markersize=markersize,linewidth=linewidth)
ax.fill_between(np.arange(2,12),np.array(e_i_low),np.array(e_i_high),alpha=0.3,color=CB_color_cycle[0])
# plt.set_xscale('log')
ax.set_xlabel(r'Size of the dictionary $\mathcal{Q}$', fontsize=xlabel_fontsize)
ax.set_ylabel('Avg. RMSE', fontsize=ylabel_fontsize)
# plt.set_xticks([0.001,0.01,0.1])
# plt.set_xticklabels(['$10^{-3}$',r'$10^{-2}$',r'$10^{-1}$'], fontsize=ticks_fontsize)
ax.set_xticks([2,4,6,8,10],[2,4,6,8,10],fontsize=ticks_fontsize)

ax.set_yticks([0,0.005,0.01,0.015], labels=['0','0.005','0.010','0.015'],fontsize=ticks_fontsize)
# plt.title(, fontsize=title_fontsize)
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels,loc='center right',fontsize=legend_fontsize)
plt.subplots_adjust(left=0.12,
                    bottom=0.15, 
                    right=0.78, 
                    top=0.9, 
                    wspace=0.25, 
                    hspace=0.5)
plt.savefig("figures/Figure_8.pdf",dpi=fig.dpi)
plt.show()
```
