```python
import matplotlib.pyplot as plt
import numpy as np

import glob
import pandas as pd

file = glob.glob('../results/Figure_8_9/*.csv')[0]
df = pd.read_csv(file)

mean_times = df['time_mean']
std_times = df['time_std']

dictionary_sizes = list(range(1,11))
fig = plt.figure(figsize=(8, 6))
plt.plot(dictionary_sizes, mean_times, 'o-', label="Mean computation time")
plt.fill_between(dictionary_sizes, np.array(mean_times) - np.array(std_times), np.array(mean_times) + np.array(std_times), alpha=0.2)
plt.xlabel(r"Size of the dictionary $\mathcal{Q}$")
plt.ylabel("Computation time (s)")
# plt.yscale("log")
plt.legend()
fig.savefig("figures/Figure_9.pdf")
```
